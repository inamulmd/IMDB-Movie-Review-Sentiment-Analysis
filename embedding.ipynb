```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
## sentences
sent=['the glass of milk',
      'the glass of juice',
      'the cup of tea',
      'I am a good boy',
      'I am a good developer',
      'understand the meaning of words',
      'your videos are good',

]
```

```python
sent
```

```python
#3 Define the vocabulary size
voc_size=10000
```

```python
# from tensorflow.keras.preprocessing.text import Tokenizer
# from tensorflow.keras.utils import pad_sequences
# from tensorflow.keras.models import Sequential
# from tensorflow.keras.layers import Embedding

# texts = ["I love AI", "AI is future"]

# tokenizer = Tokenizer()
# tokenizer.fit_on_texts(texts)

# sequences = tokenizer.texts_to_sequences(texts)

# padded = pad_sequences(sequences, maxlen=5)

# model = Sequential()
# model.add(Embedding(input_dim=1000, output_dim=64, input_length=5))
```

```python
### One hot Representation
one_hot_rep=[one_hot(words,voc_size)for words in sent]
one_hot_rep
```

```python
## word Embedding Representation
from tensorflow.keras.layers import Embedding
# Embedding(input_dim, output_dim, input_length)
# input_dim → vocabulary size  ..number of unique words your model can handle.
# output_dim → vector size (e.g., 50, 100, 300)
# input_length →   Length of each input sequence after padding

#from tesnsorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length=8
embedded_docs=pad_sequences(one_hot_rep,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
## feature representation
dim=10
```

```python
from tensorflow.keras.layers import Input
```

```python
model = Sequential()
model.add(Input(shape=(sent_length,)))
model.add(Embedding(voc_size, dim))

model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
